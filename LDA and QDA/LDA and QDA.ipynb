Linear Discriminant Analysis

```python
class LinearDiscriminantAnalysis:
    """
    Fits it on data, then uses predict to get results.
    """

    def __init__(self):
      self.classes_ = None
      self.scalings_ = None
      self.n_classes = None
    
    def fit(self, X, y):
        """Fit the training data
        Parameters
        ----------
        X : array-like, shape = [n_samples, n_features]
            Features
        y : array-like, shape = [n_samples, n_labels]
            Labels
        Returns
        -------
        self : object
        """
        n_samples, n_features = X.shape
        '''
        Write your own code
        '''
        self.classes_ = np.unique(y)
        self.n_classes = self.classes_.size

        mean = X.mean(axis=0)
        class_means = np.zeros((self.n_classes, n_features))
        for i, c in enumerate(self.classes_):
            class_means[i] = X[y == c].mean(axis=0)

        S_w = np.zeros((n_features, n_features))
        S_b = np.zeros((n_features, n_features))
        for i, c in enumerate(self.classes_):
            Xi = X[y == c]
            Xi_centered = Xi - class_means[i]
            S_w += np.dot(Xi_centered.T, Xi_centered)
            n = Xi.shape[0]
            mean_diff = class_means[i] - mean
            S_b += n * np.outer(mean_diff, mean_diff)

        A = np.dot(np.linalg.pinv(S_w), S_b)
        eigenvalues, eigenvectors = np.linalg.eig(A)
        indices = np.argsort(eigenvalues)[::-1]
        self.scalings_ = eigenvectors[:, indices][:, :self.n_classes - 1]
    
    def predict(self, X):
        """ Predicts the value after the model has been trained.
        Parameters
        ----------
        X : array-like, shape = [n_samples, n_features]
            Features
        Returns
        -------
        Predicted value
        """
        n_samples, n_features = X.shape
        '''
        Write your own code
        '''
        scores = np.dot(X, self.scalings_)
        y_pred = self.classes_[np.argmax(scores, axis=1)]
        return y_pred
```

Quadratic Discriminant Analysis (Optional) (15 bonus points)

```python
class QuadraticDiscriminantAnalysis:
    """
    Fits it on data, then uses predict to get results.
    """

    def __init__(self):
        self.classes_ = None
        self.priors = None
        self.means_ = None
        self.covariance_ = None
        self.n_classes = None
    
    def fit(self, X, y):
        """Fit the training data
        Parameters
        ----------
        X : array-like, shape = [n_samples, n_features]
            Features
        y : array-like, shape = [n_samples, n_labels]
            Labels
        Returns
        -------
        self : object
        """
        n_samples, n_features = X.shape
        '''
        Write your own code
        '''
        self.classes_ = np.unique(y)
        self.n_classes = len(self.classes_)
        self.priors_ = np.zeros(self.n_classes)
        self.means_ = np.zeros((self.n_classes, n_features))
        self.covariance_ = np.zeros((n_features, n_features))

        for i, c in enumerate(self.classes_):
          X_c = X[y == c]
          self.priors_[i] = X_c.shape[0] / float(n_samples)
          self.means_[i] = np.mean(X_c, axis=0)
          diff = X_c - self.means_[i]
          self.covariance_ += diff.T.dot(diff) / float(X_c.shape[0])
    
    def predict(self, X):
        """ Predicts the value after the model has been trained.
        Parameters
        ----------
        X : array-like, shape = [n_samples, n_features]
            Features
        Returns
        -------
        Predicted value
        """
        n_samples, n_features = X.shape
        '''
        Write your own code
        '''
        probs = np.zeros((n_samples, self.n_classes))
        for i, c in enumerate(self.classes_):
          diff = X - self.means_[i]
          temp = np.sum((diff).dot(np.linalg.pinv(self.covariance_)) * (diff), axis=1)
          probs[:, i] = np.log(self.priors_[i]) - 0.5 * temp
        y_pred = self.classes_[np.argmax(probs, axis=1)]
        return y_pred
```
